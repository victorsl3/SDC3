```python
from astropy.io import fits
from astropy.table import Table
from PIL import Image
import numpy as np
import torch
import torchvision

path_datos="/home/21953404Victor/SDC3/SDC3GIT/data/"
```

```python
path =path_datos+ "ZW2.msw_image.fits"
hdul = fits.open(path)
print(hdul.info())
```

```python
path2 = path_datos+"ZW2.msn_image.fits"
hdul2 = fits.open(path2)
print(hdul2.info())
```

```python
path_cat = path_datos + "ZW2.cat"
cat = Table.read(path_cat, format="ascii")
```

```python
# Crear un archivo de anotación
clase = "galaxia"
with open("galaxias_sextractor.txt", "w") as f:
    for row in cat:
        # Obtener las coordenadas de la galaxia
        x = row["X_IMAGE"]
        y = row["Y_IMAGE"]

        # Escribir la anotación en el archivo
        f.write(f"{x} {y} {clase}\n")
```

```python

model = torchvision.models.detection.fasterrcnn_resnet50_fpn(pretrained=True)
model.eval()
```

```python
it=0
for array in hdul[0].data:
    # Crear imagen a partir del array
    img = Image.fromarray(np.uint8(array * 255))
    # Guardar imagen como .jpg
    it=it+1
    img.save('imagen '+str(it)+'.jpg')
```

```python
# Cargar imagen
img = Image.open('image.jpg')

# Transformar imagen
transform = torchvision.transforms.Compose([
    torchvision.transforms.ToTensor()
])
img = transform(img)

# Detectar objetos en la imagen
with torch.no_grad():
    prediction = model([img])

print(prediction)
```
